This file explains how the data treatment was done for our datasets used

Transfered the column LABEL to the last column in all datasets

So, if a column in a dataset has 60% or more of its entries as N/D, empty, NULL, etc... this column got removed! muahahaha

If a column has empty values in it, less than 60% of course, I choose to apply the median value (for NUMERIC VALUES) and "Unknown" (for CATEGORIC VALUES)

THEN... BINARIZATION OF TEXT BY: ONE-HOT-ENCODING (for the columns with CATEGORICAL VALUES)

AT THIS POINT, OUR DATASET DOES NOT HAVE ANY EMPTY VALUE OR NOT-NUMERICAL VALUE

then... STANDARDIZATION, put numeric values in a form of: Standard deviation = 1, and mean = 0.

The column LABEL was not affected at all, I mean, if it was text-based classes, they became 1,2,3,etc...
