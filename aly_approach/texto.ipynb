How logistic regression works:

Fist, we need to make it clear that's a classification algorithm, and also, it's a binary classification one. How it does what it does?

This algoritm works with multiple runs of itself to minnimize the error, but it needs to start from somewhere.

1- It calculates the linear combinario z = w.x + b, where X is the value of a column and W it's weight. The 1st step uses weight 0, by it means, it basicly draws a straightline that divides out space in 50%/50%.

2- The value of Z, later is compressed by a SIGMOID funcion, because Z can value between +inf and -inf. The result of the Sigmoid funcion will be somewhere between 0 and 1, representing the probability of a line to be labeled as class "1", if the sigmoid function is: >=0,50, it's 1, and if it's < 0,50, it's class 0. So, for our first iteraction is basicly assuming all our examples is labeled as "class 1".

3- Now it's time for the LossFunction. For each data(line) we take the value of the loss and then, sum all of them and divide by the number of lines on our dataset, basicly it's the MEAN of the LOSSES. By the end, this value is called "Cost". This value "Cost" is not used for the next step, it's just a measurement of "how good is our model".

4- At the same time we calculate the LossFunction, we take the error of each line and multiply it for the value of it's inputs. for One line, we gonna have the "Gradient", it shows how much of our weight we should add or subtract to reduce our error. In multiple lines, we make the mean of the gradientes of all lines.

5- At the end, we gonna change our initial weights by adding or subtracting it by the learning rate * gradient of each one of our columns.

6- With the updated weights, we can repeat the process a lot of times to better find our equation that divides our whole space into two.
