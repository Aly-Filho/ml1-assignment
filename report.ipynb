### 2) Understand the algorithm and hypothesize which of the suggested data characteristics affects it the most.  

#### - How it works:

Deep down, Logistic Regression works by finding the optimal vector of weights **w** that maximizes the likelihood of the observed data using a process called Gradient Descent to minimize the Log-Loss function.  

Unlike linear regression, which minimizes the sum of squared errors, Logistic Regression calculates the dot product of inputs and weights to position a hyperplane in **n-dimensional** space.  

This linear output is then transformed by the Sigmoid function, creating a non-linear probability mapping where the "decision boundary" is the specific threshold—typically P=0.5—where the model transitions from predicting one class to another.  

Essentially, the "line" seen is the geometric representation of the weights effectively balancing the features to minimize the cost of misclassification across the entire training set.

#### - Limitations:

The primary limitation of Logistic Regression stems from its geometric simplicity: **a single linear equation defines a hyperplane**, which is essentially a flat "cut" through your data space. 

In a two-dimensional graph, this cut is a single straight line. By mathematical necessity, **one straight line can only bisect a continuous space into exactly two distinct regions**.  

This physical constraint dictates the algorithm's binary nature—**it can only decide if a point falls on one side of the fence or the other**.  


**It cannot "carve out" a third middle zone or isolate multiple separate clusters using a single function.**
